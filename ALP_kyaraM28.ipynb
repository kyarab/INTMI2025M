3- crie um algoritmo que receba o nome do aluno, e 3 notas. depois calcule a média das notas e imprima na tela: o aluno (aluno) ésta com (média) de média!

```python
aluno = str(input("digite seu nome"))
nota1 = float(input("digite uma nota"))
nota2 = float(input("digite sua outra nota"))
nota3 = float(input("digite a terceira nota"))

média = nota1+nota2+nota3/3

print(f"O aluno {aluno} ésta com {média:.2f} de média")
```

```python
resposta = input('Digite "S" para continuar ou "N" para sair: ').strip().upper()

if resposta == "S":
    print("Você escolheu continuar.")
elif resposta == "N":
    print("Você escolheu sair.")
else:
    print("Opção inválida.")
```

```python
numero = float(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
elif numero == 0:
    print("O número é zero.")
else:
    print("O número é negativo.")
```

```python
numero = int(input("Digite um número de 1 a 7: "))

if numero == 1:
    print("Domingo")
elif numero == 2:
    print("Segunda-feira")
elif numero == 3:
    print("Terça-feira")
elif numero == 4:
    print("Quarta-feira")
elif numero == 5:
    print("Quinta-feira")
elif numero == 6:
    print("Sexta-feira")
elif numero == 7:
    print("Sábado")
else:
    print("Número inválido! Digite um número de 1 a 7.")
```

```python
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print("O ano é bissexto.")
else:
    print("O ano não é bissexto.")
```

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)

print(f"Salário total: R$ {salario:.2f}")
```
