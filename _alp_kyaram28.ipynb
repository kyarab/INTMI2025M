```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)

print(f"Salário total: R$ {salario:.2f}")
```

```python
 print("=========================")
print("  CALCULADORA INTERATIVA ")
print("=========================")

while True:  # Loop principal
    print("\nDigite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")

    opcao = input("Opção: ")

    # ---------- OPÇÃO 0 ----------
    if opcao == "0":
        print("Saindo do programa... Obrigado por usar!")
        break

    # ---------- OPÇÃO 1 ----------
    elif opcao == "1":
        print("\n=== MÚLTIPLOS DE 5 ===")
        inicio = int(input("Digite o início do intervalo: "))
        fim = int(input("Digite o final do intervalo: "))

        print(f"Múltiplos de 5 entre {inicio} e {fim}:")
        for n in range(inicio, fim + 1):
            if n % 5 == 0:
                print(n)

    # ---------- OPÇÃO 2 ----------
    elif opcao == "2":
        print("\n=== CONTAGEM DE PARES ===")
        inicio = int(input("Digite o início do intervalo: "))
        fim = int(input("Digite o final do intervalo: "))

        pares = 0
        for n in range(inicio, fim + 1):
            if n % 2 == 0:
                pares += 1

        print(f"Quantidade de números pares no intervalo: {pares}")

    # ---------- OPÇÃO 3 ----------
    elif opcao == "3":
        print("\n=== VERIFICADOR DE NÚMERO PRIMO ===")
        num = int(input("Digite um número inteiro positivo: "))

        if num < 2:
            print("Não é primo.")
        else:
            primo = True
            for i in range(2, num):
                if num % i == 0:
                    primo = False
                    break

            if primo:
                print(f"{num} é primo.")
            else:
                print(f"{num} NÃO é primo.")

    # ---------- OPÇÃO 4 ----------
    elif opcao == "4":
        print("\n=== CONVERSOR DE TEMPERATURA ===")
        c = float(input("Digite a temperatura em Celsius: "))
        f = c * 1.8 + 32
        print(f"{c}°C equivalem a {f:.2f}°F")

    # ---------- OPÇÃO INVÁLIDA ----------
    else:
        print("Opção inválida! Tente novamente.")]
```
